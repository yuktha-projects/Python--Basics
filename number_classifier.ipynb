Objective: Demonstrate structured decision logic using nested if-else statements.

This project simulates the classification logic using numeric input.

## Problem Statement

Develop a conditional logic system that classifies a given number into one of the following categories:
- Positive Even
- Positive Odd
- Negative Even
- Negative Odd
- Zero

## Approach

Step 1: Determine whether the number is positive, negative, or zero.  
Step 2: If non-zero, determine whether the number is even or odd using modulus operator (%).  
Step 3: Print classification result.

```python
num = 5

if num > 0:
    if num % 2 == 0:
        result = "Positive Even"
    else:
        result = "Positive Odd"
elif num < 0:
    if num % 2 == 0:
        result = "Negative Even"
    else:
        result = "Negative Odd"
else:
    result = "Zero"

print("Input Number:", num)
print("Classification:", result)
```

## Edge Case Testing

```python
test_values = [0, -4, 11, -7, 9]

for num in test_values:
    if num > 0:
        if num % 2 == 0:
            result = "Positive Even"
        else:
            result = "Positive Odd"
    elif num < 0:
        if num % 2 == 0:
            result = "Negative Even"
        else:
            result = "Negative Odd"
    else:
        result = "Zero"

    print(f"{num} = {result}")
```

## Logical Flow

1. Check sign of number.
2. If number is non-zero, evaluate parity using modulus operation.
3. Assign category based on condition.
4. Print structured result.
