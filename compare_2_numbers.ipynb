###**COMPARE 2 NUMBERS USING if-else statements**



```python
a = 10
b = 20

print(f'Value of a: {a}')
print(f'Value of b: {b}')

# logic to check the edge cases
if a > b:
    print("Result: a is greater than b")
elif b > a:
    print("Result: b is greater than a")
else:
  # This handles the edge case where numbers are equal
    print("Result: Both numbers are equal")
```
